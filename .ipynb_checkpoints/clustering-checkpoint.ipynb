```python
import pandas as pd
import seaborn as sns
from palmerpenguins import load_penguins
from sklearn import preprocessing
import numpy as np
from numpy.random import uniform
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
```

```python
penguins = load_penguins()
penguins = penguins.dropna()
```

The first step is to standardize the data, i.e. ensure that each variable has a mean of 0 and a standard deviation of 1

```python
X = preprocessing.scale(penguins['flipper_length_mm'])
Y = preprocessing.scale(penguins['bill_length_mm'])

X_train = np.array(list(zip(X,Y)))
```

Define a function to calculate the distances betwwen a point and a dataset of points


```python
def euclidean(point, data):
    return np.sqrt(np.sum((point-data)**2, axis=1))
```

```python
class KMeans:
    def __init__(self, n_clusters=8, max_iter=1000):
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.all_centroids = []
        
    def fit(self, X):
        #randomly select centroid start points
        min_, max_ = np.min(X, axis=0), np.max(X, axis=0)
        self.centroids = [uniform(min_, max_) for _ in range(self.n_clusters)]
        
        iteration = 0
        prev_centroids = None
        while np.not_equal(self.centroids, prev_centroids).any() and iteration < self.max_iter:
            sorted_points = [[] for _ in range(self.n_clusters)]
            self.all_centroids.append(self.centroids)
            for x in X:
                dists = euclidean(x, self.centroids)
                centroid_idx = np.argmin(dists)
                sorted_points[centroid_idx].append(x)
            prev_centroids = self.centroids
            self.centroids = [np.mean(cluster, axis=0) for cluster in sorted_points]
            for i, centroid in enumerate(self.centroids):
                if np.isnan(centroid).any():
                    self.centroids[i] = prev_centroids[i]
            iteration +=1
        return self.all_centroids
```

```python
kmeans = KMeans(n_clusters=3)
list_centroids = kmeans.fit(X_train)
```

```python
plt.scatter(x=X, y=Y)
for centroid in list_centroids:
    plt.scatter(centroid[0], centroid[1])
```
